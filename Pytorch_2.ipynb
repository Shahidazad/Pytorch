```python
import numpy as np
import torch
```

```python
# Input (temp, rainfall, humidity)
inputs = np.array([[73, 67, 43], 
                   [91, 88, 64], 
                   [87, 134, 58], 
                   [102, 43, 37], 
                   [69, 96, 70]], dtype='float32')
```

```python
# Targets (apples, oranges)
targets = np.array([[56, 70], 
                    [81, 101], 
                    [119, 133], 
                    [22, 37], 
                    [103, 119]], dtype='float32')
```

```python
# Convert inputs and targets to tensors
inputs = torch.from_numpy(inputs)
targets = torch.from_numpy(targets)
print(inputs)
print(targets)
```

```python
# Weights and biases
w = torch.randn(2, 3, requires_grad=True)
b = torch.randn(2, requires_grad=True)
print(w)
print(b)
```

```python
def model(x):
    return x @ w.t() + b
```

```python
# Generate predictions
preds = model(inputs)
print(preds)
```

```python
# MSE loss
def mse(t1, t2):
    diff = t1 - t2
    return torch.sum(diff * diff) / diff.numel()
```

```python
# Compute loss
loss = mse(preds, targets)
print(loss)
```

```python
# Compute gradients
loss.backward()
```

```python
# Gradients for weights
print(w)
print(w.grad)  # derivative
print(b.grad)
```

```python
print(b)
print(b.grad) # 
```

```python
# adjust weight  & reset perdiction
with torch.no_grad():
  w-=w.grad  * 1e-5  # reduce the weight
  b-=b.grad  * 1e-5 
  w.grad.zero_()
  b.grad.zero_() 
```

```python


print(w)
print(b)
```

```python
# calculate loss

# loss we reduce after changing the bias weight
preds=model(inputs)
loss=mse(preds,targets)
print(loss)
```

```python
# train 1000 epochs   # this will increase the accuracy
for i in range(100):
  preds=model(inputs)
  loss=mse(preds,targets)
  loss.backward()
  with  torch.no_grad():
    w -=w.grad * 1e-5
    b -=b.grad * 1e-5
    w.grad.zero_()
    b.grad.zero_()
```

```python
preds=model(inputs)
loss=mse(preds,targets)
print(loss)
```

```python
import torch.nn  as nn 
```

```python
# Input (temp, rainfall, humidity)
inputs = np.array([[73, 67, 43], 
                   [91, 88, 64], 
                   [87, 134, 58], 
                   [102, 43, 37], 
                   [69, 96, 70], 
                   [74, 66, 43], 
                   [91, 87, 65], 
                   [88, 134, 59], 
                   [101, 44, 37], 
                   [68, 96, 71], 
                   [73, 66, 44], 
                   [92, 87, 64], 
                   [87, 135, 57], 
                   [103, 43, 36], 
                   [68, 97, 70]], 
                  dtype='float32')

# Targets (apples, oranges)
targets = np.array([[56, 70], 
                    [81, 101], 
                    [119, 133], 
                    [22, 37], 
                    [103, 119],
                    [57, 69], 
                    [80, 102], 
                    [118, 132], 
                    [21, 38], 
                    [104, 118], 
                    [57, 69], 
                    [82, 100], 
                    [118, 134], 
                    [20, 38], 
                    [102, 120]], 
                   dtype='float32')

inputs = torch.from_numpy(inputs)  # CONVT TO TENSOR
targets = torch.from_numpy(targets)
```

```python
inputs
```

```python
from  torch.utils.data import TensorDataset
```

```python
train_ds= TensorDataset(inputs,targets)
train_ds[0:3]
```

```python
from torch.utils.data  import DataLoader
```

```python
batch_size = 5
train_dl = DataLoader(train_ds,batch_size,shuffle=True)
```

```python
for xb,yb in train_dl:
  print(xb)
  print(yb)
  break
```

```python
# define model
model=nn.Linear(3,2) # 3 input 2 output(target)
print(model.weight)
model.bias
```

```python
list(model.parameters())
```

```python
# generate prediction
preds=model(inputs)
preds
```

```python
import  torch.nn.functional as F
```

```python
# define loss fuct
loss_fn=F.mse_loss
```

```python
# ?F.mse_loss  # help
```

```python
loss=loss_fn(model(inputs),targets)
loss
```

```python
# define optimizer
opt=torch.optim.SGD(model.parameters(),lr=1e-5) # short cut for gradient descent reducing weights
```

```python
# to train model
def fit(num_epochs,model,loss_fn,opt,train_dl):
  # repeat for epochs
  for epoch in range(num_epochs):
    # train batch of data 
    for xb,yb in train_dl:
      # do prediction
      pred = model(xb)
      # calculate loss
      loss=loss_fn(pred,yb)
      # compute gradient 
      loss.backward()
      # update parameters
      opt.step() #for gradient descent reducing weights
      # 5. Reset the gradients to zero
      opt.zero_grad()
    if (epoch+1) % 10 ==0:
      print('Epoch [{}/{}], Loss: {:.4f}'.format(epoch+1,num_epochs,loss.item()))
```

```python
fit(100,model,loss_fn,opt,train_dl)
```
